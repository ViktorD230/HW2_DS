```python
import urllib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

url = "https://uk.wikipedia.org/wiki/Населення_України"
url_escaped = urllib.parse.quote_plus(url, "/:?=&")

data = pd.read_html(
    url_escaped,
    match="Коефіцієнт народжуваності в регіонах України",
    encoding="utf-8",
    thousands=".",
)

data[0]
```

```python
data[0].head()
```

```python
data[0].shape
```

```python
update_data = data[0].replace(to_replace="—", value=np.nan)
update_data
```

```python
update_data.dtypes
```

```python
new_data = update_data.apply(lambda x: x.str.replace(",", "."))
df = new_data.astype(
    {
        "1950": "float64",
        "1960": "float64",
        "1970": "float64",
        "2014": "float64",
        "2019": "float64",
    }
).dtypes
df
```

```python
nan_count = (new_data.isnull().sum() / len(new_data) * 100).round(1).astype(str) + "%"
print("Відсоткова частка  пропусків значень  по кожному періоду, скаладає:")
nan_count[1:]
```

```python
df_del_last = new_data.drop([27])
df_del_last
```

```python
new_data = df_del_last.astype(
    {
        "1950": "float64",
        "1960": "float64",
        "1970": "float64",
        "1990": "float64",
        "2000": "float64",
        "2012": "float64",
        "2014": "float64",
        "2019": "float64",
    }
)

new_data = new_data.fillna(new_data.iloc[:, 1:].mean().round(1))
new_data
```

```python
aver_2019 = new_data["2019"].mean().round(1)
spisok_reg = new_data[new_data["2019"] > aver_2019]
spisok_reg[["Регіон", "2019"]]
```

```python
big_2014 = new_data["2014"].idxmax()
new_data.iloc[big_2014]["Регіон"]
```

```python
diagram_2019 = new_data[["Регіон", "2019"]].set_index("Регіон")
diagram_2019.plot(kind="bar")
plt.show()
```

```python
regions = ["Київ", "Дніпропетровська", "Севастополь"]
plt.figure(figsize=(10, 6))
for region in regions:
    plt.plot(
        new_data.columns[1:],
        new_data[new_data["Регіон"] == region].iloc[:, 1:].values.flatten(),
        label=region,
    )
plt.title("Коефіціент народжуваності в різних регіонах")
plt.xlabel("Рік")
plt.ylabel("Коефіціент народжуваності")
plt.legend()
plt.show()
```

```python
regions = ["Луганська", "Львівська"]
plt.figure(figsize=(10, 6))
for region in regions:
    plt.scatter(
        new_data.columns[1:],
        new_data[new_data["Регіон"] == region].iloc[:, 1:].values.flatten(),
        label=region,
    )
plt.title("Коефіціент народжуваності по регіонах")
plt.xlabel("Рік")
plt.ylabel("Коефіціент народжуваності")
plt.legend()
plt.show()
```
