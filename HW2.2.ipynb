```python
df_upate = developers.drop(["Специализация", "Университет", "Предметная.область", "Валюта", "cls"], axis=1)
nan_count_update = (df_upate.isnull().sum() / len(df_upate) * 100).round(1).astype(str) + "%"
nan_count_update
```

```python
nan_count_update = (df_upate.isnull().sum() / len(df_upate) * 100).round(1).astype(str) + "%"
nan_count_update
```

```python
df_new = df_upate.dropna()
df_new
```

```python
df_new.shape
```

```python
python_data = df_new[df_new["Язык.программирования"] == "Python"]

python_data
```

```python
python_data.shape
```

```python
df_group = python_data.groupby(["Должность"])
df_group.head()
```

```python
new_df = df_group.agg({"Зарплата.в.месяц": ["min", "max"]})
new_df
```

```python
def fill_avg_salary(position):
    return python_data[python_data["Должность"] == position]["Зарплата.в.месяц"].mean().round(0)


python_data.loc[:, "avg"] = python_data["Должность"].apply(fill_avg_salary)
python_data
```

```python
avg = python_data["avg"]
descr_avg = avg.describe().round(1)
descr_avg
```

```python
new_data = python_data.to_csv("new_data.csv")
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(
    x="Город",
    y="salary",
    data=python_data,
    hue="Язык.программирования",
    errorbar=None,
    palette="viridis",
)
plt.xticks(fontsize=6)
plt.title("Изменение зарплаты по городам для языка программирования Python")
plt.xlabel("Город")
plt.ylabel("Зарплата в месяц")
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.countplot(x="Язык.программирования", hue="Образование", data=df_new, palette="Set2")
plt.xticks(fontsize=6)
plt.title("Зависимость языков программирования от уровня образования")
plt.xlabel("Язык программирования")
plt.ylabel("Количество")
plt.show()
```
